```python
import psycopg2 
import os
import sys

def isprime(num):
    count = 0
    for i in range(1,num+1):
        if num % i == 0:
            count+=1
    if (count == 2):
        return True
    else:
        return False

conn = None
try:
    conn = psycopg2.connect(database='flis_db', user='postgres',password='Sidhu237@',host='localhost',port='5432')
    # Create a new cursor
    cur = conn.cursor()
    cur.execute("select p.name, t.name, p.jersey_no from players p, teams t where p.team_id=t.team_id order by p.name desc, t.name desc")
    rows = cur.fetchall() # fetch all rows
    
    for row in rows:
        pname = row[0]
        tname = row[1]
        jno = row[2]
        
        if isprime(jno):
            print(pname+", "+tname)
    # close the cursor
    cur.close()

except(Exception, psycopg2.DatabaseError) as error:
    print(error)
finally:
    if conn is not None:
        conn.close()
```

```python
import psycopg2, sys, os
f = open('name.txt','r')
name = f.read()
conn = None
conn = psycopg2.connect(database='lis_db',user='postgres',password='Sidhu237@',host='localhost',port='5432')
cur = conn.cursor()
cur.execute('''select s.student_fname,d.department_name,s.dob 
               from students s, departments d 
               where s.department_code=d.department_code and student_fname=%s''',(name,))
result = cur.fetchall()
#print(result)
for i in result:
    name = i[0]
    department = i[1]
    dob = str(i[2])
    year =int(dob[:4])
    if year % 2 == 0:
        is_even ='Even'
        print(f"{name},{department},{is_even}")
    else:
        is_odd ='Odd'
        print(f"{name},{department},{is_odd}")
cur.close()
conn.close()
```

```python
import psycopg2, sys, os
f = open('date.txt','r')
date = f.read()

conn = None
conn = psycopg2.connect(database='flis_db',user='postgres',password='Sidhu237@',host='localhost',port='5432')
cur = conn.cursor()
cur.execute('''select r.name   
               from referees r join match_referees mr on r.referee_id=mr.referee join matches m on mr.match_num=m.match_num
               where match_date=%s''',(date,))
result=cur.fetchall()
for i in result:
    a=i[0]
    b=a.split()
    s=b[-1]
    for j in range(len(b)-1):
        w=b[j][0]
        s=s+" "+w+"."
    print(s)
cur.close()
conn.close()
```
